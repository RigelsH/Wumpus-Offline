```python
import sys
sys.path.append('../')
import tracemalloc
import glob
import json
import wumpus
from astar import AstarPlayer
from collections import defaultdict
```

```python
files = glob.glob("../data/*")

result = defaultdict(list)

for file in sorted(files):
    tracemalloc.start()   # start tracing memory allocations

    with open(file) as f:
        world_json = json.load(f)

        episode_result = wumpus.run_episode(
            world=wumpus.WumpusWorld.from_JSON(world_json), 
            player=AstarPlayer(), 
            show=False, 
            horizon=1000
        )

    current, peak = tracemalloc.get_traced_memory()   # get current and peak memory usage
    tracemalloc.stop()
    memory_usage = peak / (1024)
    formatted_memory = "{:.2f} KB".format(memory_usage)  

    result[file].append((formatted_memory)) 
```

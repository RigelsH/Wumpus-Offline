```python
import sys
sys.path.append("../")
import glob
from collections import defaultdict
import json
import wumpus
from bfs import BfsPlayer
import time
```

```python
files = glob.glob("../data/*")

result = defaultdict(list)

for file in sorted(files):
    start_time = time.perf_counter()    
    
    with open(file) as f:
        world_json = json.load(f)
        
        episode_result = wumpus.run_episode(
            world=wumpus.WumpusWorld.from_JSON(world_json), 
            player=BfsPlayer(), 
            show=False, 
            horizon=1000
        )
        
    end_time = time.perf_counter()    
    elapsed_time = (end_time - start_time)*1000  
    formatted_time = "{:.2f} ms".format(elapsed_time)
    
    result[file].append(formatted_time)   
    
```
